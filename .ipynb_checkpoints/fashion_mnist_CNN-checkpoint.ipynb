# CNN sur FASHION_MNIST
##### FASHION_MNIST est un ensemble de données classiques

Fashion-MNIST est un ensemble de données d'images d'articles de Zalando composé d'un ensemble d'apprentissage de 60 000 exemples et d'un ensemble de test de 10 000 exemples. Chaque exemple est une image en niveaux de gris 28x28, associée à une étiquette de 10 classes.

<img src="fashion-mnist-sprite.png"/>

### Informations


>Objectif est de prédire a partir d'une image d'articles la catégorie auxquelles elle appartient

```python
#Importation des modules
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Importation du dataSet de Mnist
from tensorflow.keras.datasets import fashion_mnist
(X_train,y_train),(X_test,y_test)=fashion_mnist.load_data()
```

```python
#Dimension données entrainement
X_train.shape
```

```python
#Dimension données test
X_train.shape
```

```python
import random
#Visulalisation des données entrainements
plt.figure(figsize=(15,15))
for i in range(9):
    
    plt.subplot(3,3,i+1)
    plt.imshow(X_train[random.randint(0,60000)])
```

```python
import random
#Visulalisation des données entrainements
plt.figure(figsize=(15,15))
for i in range(9):
    
    plt.subplot(3,3,i+1)
    plt.imshow(X_test[random.randint(0,10000)])
```

#### Remarque
> C'est un probleme de classification.
> On doit transformer les targets y en données catégorielles

```python
from tensorflow.keras.utils import to_categorical
y_cat_train=to_categorical(y_train)
y_cat_test=to_categorical(y_test)
```

```python
#Normalisation des images
X_train=X_train/255
X_test=X_test/255
```

### Remarque
> FASHION_MNIST est un enseble d'image noir blanc de taille 28,28.Soit 28,28,1 avec 1 le canal de couleur 

```python
X_train=X_train.reshape(60000,28,28,1)
X_test=X_test.reshape(10000,28,28,1)
```

## Création du modele

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,MaxPooling2D,Flatten
```

```python
model=Sequential()
model.add(Conv2D(filters=32,kernel_size=(3,3),strides=(1,1),input_shape=(28,28,1),activation='relu'))
model.add(MaxPooling2D(pool_size=(1,1)))
model.add(Conv2D(filters=64,kernel_size=(3,3),strides=(1,1),input_shape=(28,28,1),activation='relu'))
model.add(MaxPooling2D(pool_size=(1,1)))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
model.compile(loss="binary_crossentropy",optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
#Eviter les problemes Overfitting
early=EarlyStopping(monitor="val_loss",patience=2)
```

```python
model.summary()
```

```python
#Entrainement
model.fit(X_train,y_cat_train,validation_data=(X_test,y_cat_test),epochs=1000,callbacks=[early])
```

## Evaluation de la performance

```python
losses=pd.DataFrame(model.history.history)
losses.plot()
```

```python
predictions=np.argmax(model.predict(X_test),axis=1)
predictions
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
confusion_matrix(y_test,predictions)
```

```python
#Visuatlisation de la matrice de confusion
plt.figure(figsize=(10,10))
sns.heatmap(confusion_matrix(y_test,predictions),annot=True)
```
